```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Sample training data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # 2D array for sklearn
Y = np.array([2, 4, 6, 8, 10])

# Step 2: Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Step 3: Get the parameters
m = model.coef_[0]      # Slope
c = model.intercept_    # Intercept

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Step 4: Predict the output
Y_pred = model.predict(X)

# Step 5: Plot the result
plt.scatter(X, Y, color='blue', label='Actual')
plt.plot(X, Y_pred, color='red', label='Predicted Line')
plt.xlabel('X (input)')
plt.ylabel('Y (target)')
plt.title('Linear Regression using Scikit-learn')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Sample training data (X: input, Y: output)
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 6, 8, 10])  # Ideal linear: y = 2x

# Step 2: Initialize parameters
m = 0  # slope (2)
c = 0  # intercept (1)
alpha = 0.01  # learning rate
epochs = 1000  # number of iterations
n = len(X)

# Step 3: Gradient Descent Loop
for i in range(epochs):
    Y_pred = m * X + c  # forward propagation (hypothesis)
    error = Y_pred - Y

    # Cost function (Mean Squared Error)
    cost = (1/n) * np.sum(error ** 2)

    # Gradients (derivatives)
    dm = (2/n) * np.sum(error * X)
    dc = (2/n) * np.sum(error)

    # Parameter updates
    m = m - alpha * dm
    c = c - alpha * dc

    # Print every 100th iteration
    if i % 100 == 0:
        print(f"Epoch {i}: Cost = {cost:.4f}, m = {m:.4f}, c = {c:.4f}")

# Step 4: Final Output
print("\nFinal parameters:")
print(f"Slope (m): {m:.4f}")
print(f"Intercept (c): {c:.4f}")

# Step 5: Plot the result
plt.scatter(X, Y, color='blue', label='Actual')
plt.plot(X, m*X + c, color='red', label='Predicted Line')
plt.xlabel('X (input)')
plt.ylabel('Y (target)')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

# Data
X = np.array([[1], [2], [4], [5]])
y = np.array([0, 0, 1, 1])

# Train model
model = LogisticRegression()
model.fit(X, y)

# Predict probability
print("Prob of Passing (3 hrs):", model.predict_proba([[3]])[0][1])

# Predict class
print("Predicted class (3 hrs):", model.predict([[3]]))
```
